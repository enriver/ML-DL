## 이미지 흐리게 하기

이미지를 흐리게 하기 위해서 각 픽셀을 주변 픽셀의 평균값으로 변환한다.

주변 픽셀에 수행되는 연산을 커널(kernel)이라 한다.

커널의 크기는 흐림의 정도를 결정하고, 커널이 클수록 이미지가 더 부드러워진다.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# 이미지 흐리게 하기

image_blurry=cv2.blur(image,(50,50))
```

```python
plt.imshow(image_blurry, cmap='gray'), plt.axis('off')
plt.show()
```

```python
# 커널의 크기 100*100

image_very_blurry=cv2.blur(image,(200,200))
```

```python
plt.imshow(image_very_blurry, cmap='gray'), plt.axis('off')
plt.show()
```

```python
# 커널 생성

kernel=np.ones((5,5)) / 500.0
```

```python
kernel
```

```python
# 커널 적용

image_kernel=cv2.filter2D(image, -1 , kernel)
```

```python
plt.imshow(image_kernel, cmap='gray'), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
# 가우시안 블러

image_very_blurry=cv2.GaussianBlur(image, (3,3), 0)
```

```python
plt.imshow(image_very_blurry, cmap='gray'), plt.xticks([]), plt.yticks([])
plt.show()
```

GaussianBlur 함수의 세번째 매개변수는 x축 방향의 표준편차 이다.

0으로 지정하면 ((너비-1)*0.5 -1)*0.3 + 0.8 과 같이 계산한다.

Y축 방향의 표준편차는 기본값이 0이다.

가우시안 블러에 사용한 커널은 각 축 방향으로 가우시안 분포를 따르는 1차원 배열을 만든다음 외적하여 생성한다.

getGaussianKernel 함수를 사용하여 1차원 배열을 생성하고, numpy 의 outer 함수로 외적을 계산할 수 있다

```python
gaus_vector=cv2.getGaussianKernel(5,0)
```

```python
gaus_vector
```

```python
# 벡터를 외적하여 커널 생성

gaus_kernel=np.outer(gaus_vector,gaus_vector)
```

```python
gaus_kernel
```

```python
# 커널 적용

image_kernel=cv2.filter2D(image, -1, gaus_kernel)
```

```python
plt.imshow(image_kernel, cmap='gray'), plt.xticks([]), plt.yticks([])
plt.show()
```

## 이미지 이진화하기

이미지 이진화(임계처리 - thresholding)는 어떤 값보다 큰 값을 가진 픽셀을 흰색으로 만들고,

작은 값을 가진 픽셀은 검은색으로 만드는 과정이다.

적응적 이진화(adaptive thresholding)는 주변 픽셀의 강도로부터 임계값을 결정한다.

이미지 안의 영역마다 빛 조건이 달라질 때 도움이 된다.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image_gray=cv2.imread('c:/Users/river/Desktop/ADP/plane.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# 적응적 이진화 적용

max_output_value=255
neighborhood_size=99
subtract_from_mean=10

image_binarized=cv2.adaptiveThreshold(image_gray, max_output_value, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, neighborhood_size, subtract_from_mean)
```

```python
plt.imshow(image_binarized, cmap='gray'), plt.axis('off')
plt.show()
```

adaptiveTrheshold 함수에는 네 개의 중요한 매개변수가 있다.

- max_output_value 는 출력 픽셀 강도의 최대값을 결정한다.
- cv2.ADAPTIVE_THRESH_GAUSSIAN_C 는 픽셀의 임계값을 주변 픽셀 강도의 가중치 합으로 설정한다.
- cv2.ADAPTIVE_THRESH_MEAN_C 로 주변 픽셀의 평균을 임계값으로 설정할 수 있다.
- 마지막 두개의 매개변수는 블록 크기(픽셀의 임계값 결정에 사용하는 주변 영역 크기) 와
- 계산된 임계값에서 뺄 상수 (임계값을 수동으로 미세 조정하는데 사용하는 값) 이다

```python
# cv2.ADAPTIVE_THRESH_MEAN_C

image_mean_threshold=cv2.adaptiveThreshold(image_gray, max_output_value, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, neighborhood_size, subtract_from_mean)
```

```python
plt.imshow(image_mean_threshold, cmap='gray'), plt.axis('off')
plt.show()
```

## 배경 제거하기
#### 원하는 전경 주위에 사각형 박스를 그리고 그랩컷(grabCut) 알고리즘 실행

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image_bgr=cv2.imread('c:/Users/river/Desktop/ADP/plane_256x256.jpg')
image_rgb=cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
```

```python
# 사각형 좌표 : 시작점의 x, 시작점의 y, 너비(width), 높이(height)

rectangle=(0,56,256,150)
```

```python
# 초기 마스크 설정

mask=np.zeros(image_rgb.shape[:2],np.uint8)
```

```python
# grabCut 에 사용할 임시 배열 생성

bgdModel=np.zeros((1,65), np.float64)
fgdModel=np.zeros((1,65), np.float64)
```

```python
# grabCut 실행

cv2.grabCut(image_rgb, # 원본 이미지
           mask, #마스크
           rectangle, #사각형
           bgdModel, #배경을 위한 임시 배열
           fgdModel, #전경을 위한 임시 배열
           5, #반복 횟수
           cv2.GC_INIT_WITH_RECT) # 사각형을 사용한 초기화
```

```python
# 배경인 곳은 0, 그 외에는 1로 설정한 마스크 생성

mask_2=np.where((mask==2)|(mask==0), 0, 1).astype('uint8')
```

```python
# 이미지에 새로운 마스크를 곱해 배경을 제외

image_rgb_nobg=image_rgb*mask_2[:,:,np.newaxis]
```

```python
plt.imshow(image_rgb_nobg), plt.axis('off')
plt.show()
```

그랩컷이 잘 작동하더라도 이미지에 배경이 남을 수 있다.

수동으로 배경을 고칠 수 있지만, 이미지가 많을 경우 전부 고치는 것은 불가능하기 때문에

이미지에 일부 배경이 남아있는 것을 수용하는 편이 좋다.

### 그랩컷 설명

먼저 전경이 들어있는 영역 주위를 사각형으로 표시한다.

## 경계선 감지하기
#### 캐니(Canny) 경계선 감자기 사용

```python
image_gray=cv2.imread('c:/Users/river/Desktop/ADP/plane_256x256.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# 픽셀 강도의 중간값 계산

median_intensity=np.median(image_gray)
```

```python
# 중간 픽셀 강도에서 위아래 1 표준편차 떨어진 값을 임계값으로 지정

lower_threshold=int(max(0,(1.0-0.33)*median_intensity))
upper_threshold=int(min(255,(1.0+0.33)*median_intensity))
```

```python
# 캐니 경계선 감지기 적용

image_canny=cv2.Canny(image_gray, lower_threshold, upper_threshold)
```

```python
plt.imshow(image_canny, cmap='gray'), plt.axis('off')
plt.show()
```

캐니 경계선 감지기는 gradient 임계값의 저점과 고점을 나타내는 두 매개변수가 필요하다.

낮은 임계값과 높은 임계값 사이의 가능성 있는 경계선 픽셀은 약한 경계선 픽셀로 간주한다.

높은 임계값보다 큰 픽셀은 강한 경계선 픽셀로 간주한다.

## 모서리 감지하기
#### 해리스(Harris) 모서리 감지의 OpenCV 구현인 cornerHarris 사용

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image_bgr=cv2.imread('c:/Users/river/Desktop/ADP/plane_256x256.jpg')
image_gray=cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)
image_gray=np.float32(image_gray)
```

```python
# 모서리 감지 매개변수 설정

block_size=2
aperture=29
free_parameter=0.04
```

```python
# 모서리 감지

detector_responses=cv2.cornerHarris(image_gray, block_size, aperture, free_parameter)
```

```python
# 모서리 표시 부각

detector_responses=cv2.dilate(detector_responses, None)
```

```python
# 임계값보다 큰 감지 결과만 남기고 흰색으로 표시

threshold=0.2
image_bgr[detector_responses> threshold*detector_responses.max()]=[255,255,255]
```

```python
# 흑백으로 변환

image_gray=cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(image_gray, cmap='gray'), plt.axis('off')
plt.show()
```

#### cornerHarris 는 모서리 감지에 사용하는 중요한 매개변수 세 개를 가진다

1) block_size는 각 픽셀에서 모서리 감지에 사용되는 이웃 픽셀 크기이다.

2) aperture는 사용한느 소벨 커널 크기이다.

3) free_parameter는 값이 클수록 부드러운 모서리를 구별한다.

```python
# 만들어진 결과는 가능성 있는 모서리를 나타내는 흑백 이미지이다.

plt.imshow(detector_responses, cmap='gray'), plt.axis('off')
plt.show()
```

그 다음 임계값을 적용하여 가장 가능성이 높은 모서리만 남긴다.

#### goodFeaturesToTrack 모서리감지기 (Shi-Tomasi) 를 사용하여 뚜렷하게 나타난 모서리를 지정된 개수만큼 찾아낼 수 있다

주요 매개변수로는 감지할 모서리개수, 모서리가 될 최소 품질(0~1 사이), 모서리 사이의 최소 유클리드 거리 등이 있다.

```python
image_bgr=cv2.imread('c:/Users/river/Desktop/ADP/plane_256x256.jpg')
```

```python
image_gray=cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)
```

```python
# 감지할 모서리 개수

corners_to_detect=10
minimum_quality_score=0.05
minimum_distance=25
```

```python
# 모서리 감지

corners=cv2.goodFeaturesToTrack(image_gray, corners_to_detect, minimum_quality_score, minimum_distance)

corners-np.float32(corners)
```

```python
# 모서리마다 흰 원 그리기

for corner in corners:
    x, y = corner[0]
    cv2.circle(image_bgr, (x,y), 10 , (255,255,255), -1)
```

```python
# 흑백 이미지로 변환

image_rgb=cv2.cvtColor(image_bgr, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(image_rgb, cmap='gray'), plt.axis('off')
plt.show()
```

## 평균 색을 특성으로 인코딩하기

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image_bgr=cv2.imread('c:/Users/river/Desktop/ADP/plane_256x256.jpg', cv2.IMREAD_COLOR)
```

```python
# 각 채널(RGB)의 평균값 계산

channels=cv2.mean(image_bgr)
```

```python
# BGR -> RGB

observation=np.array([(channels[2],channels[1],channels[0])])
```

```python
# 채널 평균값 확인

observation
```

```python
plt.imshow(observation), plt.axis('off')
plt.show()
```

## 컬러 히스토그램을 특성으로 인코딩하기

```python
# RGB로 변환

image_rgb=cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)
```

```python
# 특성값을 담은 리스트 생성

features=[]
```

```python
# 각 컬러 채널에 대해 히스토그램 계산

colors=('r','g','b')
```

```python
# 각 채널을 반복하면서 히스토그램을 계산하고 리스트에 추가

for i, channel in enumerate(colors):
    histogram=cv2.calcHist([image_rgb], #이미지
                          [i], # 채널 인덱스
                          None, # 마스크 없음
                          [256], # 히스토그램 크기
                          [0,256]) # 범위
    features.extend(histogram)
```

```python
# 샘플의 특성값으로 벡터 생성

observation=np.array(features).flatten()
```

```python
# 처음 다섯개의 특성 출력

observation[0:5]
```

```python
# RGB 채널값 확인

image_rgb[0,0]
```

```python
'''
히스토그램은 데이터에서 값의 분포를 나타낸다
'''

import pandas as pd

data=pd.Series([1,1,2,2,3,3,3,4,5]) # 예시
```

```python
data.hist(grid=False)
plt.show()
```

```python
# 각 컬러 채널에 대해서도 히스토그램을 계산하며 그래프를 그릴 수 있다.

for i, channel in enumerate(colors):
    histogram=cv2.calcHist([image_rgb], # 이미지
                          [i], # 채널 인덱스
                          None, # 마스크 없음
                          [256], # 히스토그램 크기
                          [0,256]) # 범위
    plt.plot(histogram, color=channel)
    plt.xlim([0,256]) # x축 범위
```

히스토그램에서 볼 수 있듯이, 모든 픽셀은 거의 0~180 사이의 파랑 채널값은 가지고 있지 않고,

주로 190~210 사이의 파랑 채널값을 가지고 있음을 확인할 수 있다.
