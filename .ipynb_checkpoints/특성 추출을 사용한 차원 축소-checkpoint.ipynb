## 주성분을 사용해 특성 줄이기
#### 일련의 특성이 주어졌을 때 데이터의 분산을 유지하면서 특성의 수 줄이기

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn import datasets
```

```python
# 데이터 로드

digits=datasets.load_digits()
```

```python
# 특성행렬 표준화

features=StandardScaler().fit_transform(digits.data)
```

```python
# 99%의 분산을 유지하도록 PCA 클래스 객체 생성

pca=PCA(n_components=0.99, whiten=True)
```

```python
# PCA 수행

features_pca=pca.fit_transform(features)
```

```python
# 결과 확인

print('원복 특성 개수:',features.shape[1])
print('줄어든 특성 개수:', features_pca.shape[1])
```

n_components 를 0과 1사이로 지정하면 pca는 해당 비율의 분산을 유지할 수 있는 최소한의 특성 개수를 반환한다.

whiten=True 로 지정하면 각 주성분의 값을 평균이 0이고 분산이 1이 되도록 변환한다.

Whitening 은 주성분에 투영된 특성의 스케일을 맞추는 역할을 한다. pca는 평균을 0으로 맞추기 때문에

화이트닝 옵션 대신 나중에 투영된 특성을 표준화해도 된다.

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(features_pca[:,0],features_pca[:,1])
plt.show()
```

화이트닝되었기 때문에 두 특성의 스케일 (x,y 스케일)이 비슷하다.

PCA 클래스의 whiten 매개변수의 기본값은 False이다.

화이트닝을 적용하지 않으면 평균은 0이지만 스케일은 맞춰지지 않는다.

```python
pca_nowhiten=PCA(n_components=0.99)
features_nowhiten=pca_nowhiten.fit_transform(features)

plt.scatter(features_nowhiten[:,0], features_nowhiten[:,1])
plt.show()
```

```python
pca_nowhiten.components_.shape
```

```python
pca_nowhiten.components_
```

특성 행렬을 주성분에 투영하려면 components_  배열을 전치하여 점곱하면 된다.

```python
import numpy as np
```

```python
np.allclose(features_nowhiten, np.dot(features,pca_nowhiten.components_.T))
```

적절한 분산 비율을 선택하기 위해 전체 주성분의 설명된 분산(explained variance)에서 유지되는 분산의 양이 크게 늘어나지 않는 지점을 찾을 수 있다

n_components 매개변수를 지정하지 않으면 특성 개수만큼 주성분이 만들어진다. 주성분에 의해 설명된 분산은 explained_variance_ratio_ 속성에 저장된다

```python
pca=PCA(whiten=True).fit(features)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.show()
```

넘파이의 cumsum 함수(누적 합)를 사용하여 분산을 누적하여 그래프를 그렸다.

여기서는 대략 30개의 주성분으로도 80% 이상의 분산을 유지한다.

### - 올바르게 분산을 계산하기 위해서는 원본 데이터를 그대로 PCA에 주입하는 것이 좋다

```python
# 표준화하지 않은 원본 데이터 사용

pca.fit(digits.data)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.show()
```

## 선형적으로 구분되지 않은 데이터의 차원 축소하기

```python
from sklearn.decomposition import KernelPCA
from sklearn.datasets import make_circles
```

```python
# 선형적으로 구분되지 않는 데이터 생성

features, _ =make_circles(n_samples=1000, random_state=1, noise=0.1, factor=0.1)
```

```python
# 방사 기저 함수(radius basis function, RBF) 를 사용하여 커널 PCA 적용

kpca=KernelPCA(kernel='rbf', gamma=15, n_components=1)
features_kpca=kpca.fit_transform(features)
```

```python
print('원본 특성 개수:', features.shape[1])
print('줄어든 특성 개수:',features_kpca.shape[1])
```

커널 함수는 선형적으로 구분되지 않는 데이터를 선형적으로 구분되는 고차원으로 투영시켜 준다.

널리 사용되는 커널은 가우시안 방사 기저 함수 커널인 rbf이다.

다른 함수로는 다항식 커널(poly), 시그모이드 커널(sigmoid), 선형 투영(linear)가 있다. (linear는 표준 PCA와 동일한 결과를 보인다)

커널 PCA의 단점은 설정할 매개변수가 많다는 것이다. 값을 설정할 때는 시행착오를 거칠 수 밖에 없다는 단점이 있다.

커널 트릭은 실제 고차원으로 데이터를 변환하지 않으면서 고차원 데이터를 다루는 듯한 효과를 낸다.

## 클래스 분리를 최대화하여 특성 줄이기
#### 분류 모델에 사용될 특성 줄이기

```python
# 선형 판별 분석(LInear Discriminant Analysis - LDA) 를 사용하여 클래스를 최대한 분리하는 성분 축으로 특성 투영

from sklearn import datasets
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
# 붓꽃 데이터셋 로드

iris=datasets.load_iris()
features=iris.data
target=iris.target
```

```python
# LDA 객체를 만들고 실행하여 특성 변환

lda=LinearDiscriminantAnalysis(n_components=1) # 원하는 특성의 개수
features_lda=lda.fit(features, target).transform(features)
```

```python
# 특성 개수 출력

print('원본 특성 개수:', features.shape[1])
print('줄어든 특성 개수:', features_lda.shape[1])
```

```python
'''
explained_variance_ratio_ 를 사용하여 각 성분이 설명하는 분산의 양 확인
'''

lda.explained_variance_ratio_
```

LDA 는 분류 알고리즘이지만 차원 축소에도 자주 사용되는 기법이다.

LDA는 특성 공간을 저차원 공간으로 투영한다는 점에서 주성분 분석(PCA)와 비슷하다.

PCA가 데이터에서 분산이 최대인 성분 축에만 관심이 있는 반면, LDA는 클래스 간의 차이를 최대화하는 추가적인 목적을 가집니다.

필요한 n_components의 값을 알기 위해서는 만들어진 각 특성이 설명하는 분산을 크기순으로 정렬한 explained_variance_ratio_를 참고할 수 있다.

특히 n_components 를 None 으로 지정하여 LinearDiscriminantAnalysis 를 실행할 수 있는데, 모든 성분 특성에 의해 설명된 분산의 비율을 반환한다.

그다음 설명된 분산의 임계값을 넘기 위해 필요한 성분 개수를 계산한다. (주로 0.95 or 0.99)

```python
# LDA 생성 및 실행

lda=LinearDiscriminantAnalysis(n_components=None)
features_lda=lda.fit(features,target)
```

```python
# 설명된 분산의 비율이 담긴 배열 저장

lda_var_ratios=lda.explained_variance_ratio_ # 크기 순으로 정렬 (내림차순)
lda_var_ratios
```

```python
# 성분의 개수를 반환하는 함수

def select_n_components(var_ratio, goal_var: float) -> int:
    # 설명된 분산의 초기값 지정
    total_variance=0.0
    
    # 특성 개수의 초기값 지정
    n_components=0
    
    # 각 특성의 설명도니 분산 순회
    for explained_variance in var_ratio:
        
        # 설명된 분산 값 누적
        total_variance+=explained_variance
        
        # 성분 개수 카운트
        n_components+=1
        
        # 설명된 분산이 목표치에 도달하면
        if total_variance >= goal_var:
            break
            
    return n_components
```

```python
select_n_components(lda_var_ratios, 0.95)
```

## 행렬 분해를 사용하여 특성 줄이기
#### 음수가 아닌 특성 행렬이 있을 때 차원 축소하기

```python
from sklearn.decomposition import NMF # 비음수 행렬 분해 (non-negative matrix factorization - NMF)
from sklearn import datasets
```

```python
digits=datasets.load_digits()
```

```python
features=digits.data
```

```python
# NMF 를 만들고 학습, 적용

nmf=NMF(n_components=10, random_state=1)
features_nmf=nmf.fit_transform(features)
```

```python
# 결과 출력

print('원본 특성 개수: ',features.shape[1])
print('줄어든 특성 개수: ',features_nmf.shape[1])
```

```python
nmf.components_.shape
```

```python
# 성분의 값은 모두 양수이다

np.all(nmf.components_>=0)
```

```python
'''
원본 데이터를 복원하려면 변환된 행렬 W와 행렬 H를 점곱한다.
완벽하게 복원하지는 못하지만 두 행렬의 차이가 크지 않음을 알 수 있다
'''

np.mean(features - np.dot(features_nmf, nmf.components_))
```

## 희소한 데이터의 특성 줄이기

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import TruncatedSVD
from scipy.sparse import csr_matrix
from sklearn import datasets
import numpy as np
```

```python
datasets=datasets.load_digits()
```

```python
features=StandardScaler().fit_transform(digits.data)
```

```python
# 희소행렬 생성

features_sparse=csr_matrix(features)
```

```python
features_sparse.shape
```

```python
# TSVD 객체 생성

tsvd=TruncatedSVD(n_components=10)
```

```python
# 희소 행렬에 TSVD 적용

features_sparse_tsvd=tsvd.fit(features_sparse).transform(features_sparse)
```

```python
print('원본 특성 개수:', features_sparse.shape[1])
print('줄어든 특성 개수:', features_sparse_tsvd.shape[1])
```

TSVD는 PCA와 비슷하다.

PCA의 단계 중 하나에서 종종 기존 SVD 방식을 사용한다.

기본 SVD에서 d개의 특성이 주어지면 SVD는 d*d 크기의 분해 행렬을 만든다.

반면 TSVD는 n*n 크기의 행렬을 만든다. n은 사전에 매개변수에서 지정한 값이다.

TSVD의 이점은 PCA와 달리 희소 특성 행렬에 사용할 수 있다는 것이다.

TSVD의 이슈는 난수 생성기를 사용하기 때문에 출력 부호가 훈련하는 사이에 뒤집힐 수 있다는 것이다.

간단한 해결방법은 전처리 파이프라인마다 한 번만 fit 메소드를 호출하는 것이다. 그 다음 여러번 transform 메소드를 사용한다.

선형 판별 분석(LDA) 처럼 n_components 매개변수를 사용하여 필요한 특성의 개수를 지정한다.

최적의 성분 개수는 n_compontents를 하이퍼파라미터로 모델 선택 과정에서 최적화 하는 것이다.

또는 TSVD가 성분마다 원본 특성 행렬의 설명된 분산 비율을 제공하기 때문에 필요한 분산의 양을 설명할 수 있는 성분 개수를 선택할 수 있다.

```python
# 처음 세 개의 성분이 설명하는 분산의 비율 합

tsvd.explained_variance_ratio_[0:3].sum()
```

```python
'''
원본 특성 개수보다 하나 작게 n_components 를 지정하고 TSVD를 실행하여
원하는 원본 데이터의 분산에서 설명된 양에 맞는 성분 개수를 계산하는 함수를 만들어 이 과정을 자동화 할 수 있다
'''

tsvd=TruncatedSVD(n_components=features_sparse.shape[1]-1)
features_tsvd=tsvd.fit(features)
```

```python
# 설명된 분산을 리스트에 저장

tsvd_var_ratios=tsvd.explained_variance_ratio_
```

```python
def select_n_components(var_ratio, goal_var):
    # 설명된 분산 초기화
    total_variance=0.0
    
    # 특성 개수 초기화
    n_components=0
    
    # 특성의 설명된 분산 순회
    for explained_variance in var_ratio:
        
        #설명된 분산 누적
        total_variance+=explained_variance
        
        #성분 개수 카운트
        n_components+=1
        
        # 설명된 분산의 목표에 도달하면 반복 마침
        if total_variance>=goal_var:
            break
            
    return n_components
```

```python
select_n_components(tsvd_var_ratios,0.95)
```

자연어 처리에서는 TSVD를 잠재 의미 분석(Latent Semantic Analysis - LSA) 라고도 부른다.

PCA는 최대 분산의 방향을 찾기 위해 원점에 맞춘 특성 행렬의 공분산 행렬에서 고유 벡터를 찾는다.

이는 특성 행렬을 특이값 분해(SVD) 하여 얻은 특이 벡터와 같다.

따라서 행렬을 원점에 맞추고 TSVD를 적용하면 PCA와 거의 같은 결과가 만들어진다.

```python
features=digits.data - np.mean(digits.data, axis=0)
```

```python
pca=PCA(n_components=40, random_state=1)
features_pca=pca.fit_transform(features)
```

```python
tsvd=TruncatedSVD(n_components=40, random_state=1)
features_tsvd=tsvd.fit_transform(features)
```

```python
np.max(np.abs(features_pca-features_tsvd))
```
