#### 이상치 제거

```python
def find_outliers(x):
    Q1=np.percentile(x,25)
    Q3=np.percentile(x,75)
    
    IQR=Q3-Q1
    lower=Q1-1.5*IQR
    upper=Q3+1.5*IQR
    
    return np.where((x>upper)|(x<lower))
```

```python
# boxplot 상 IQR 외부에 있던 변수들의 이상치 제거

outlier_index=set()

for i in numeric:
    outlier_index.update(find_outliers(df[i])[0][:])
    
outlier_index=list(outlier_index)
```

```python
new_df=df.drop(outlier_index).reset_index()
```

```python
del new_df['index']
```

```python
new_df.shape
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(new_df.corr(), annot=True)
```

다른 변수에 대해서도 동일한 상관계수를 가지고 있음을 볼 수있다.
