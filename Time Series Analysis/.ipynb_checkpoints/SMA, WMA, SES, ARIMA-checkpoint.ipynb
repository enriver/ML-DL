## SMA, WMA, SES, ARIMA

- Simple Moving Average

- Weighted Moving Average

- Simple Exponential Smoothing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
airline=\
pd.read_csv('./data/airline_passengers.csv', index_col='Month', parse_dates=True)
```

```python
airline.head()
```

### Simple Moving Average

```python
airline['6M SMA']=airline['Thousands of Passengers'].rolling(window=6).mean()
airline['12m SMA']=airline['Thousands of Passengers'].rolling(window=12).mean()
```

```python
airline.head(15)
```

```python
airline.plot(figsize=(15,8))
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(airline['Thousands of Passengers'],airline['6M SMA'].fillna(method='bfill'))
```

### Weighted Moving Average

- exponential weighted moving average

```python
airline['WMA12']=airline['Thousands of Passengers'].ewm(span=12).mean()
airline['WMA6']=airline['Thousands of Passengers'].ewm(span=6).mean()
```

```python
airline[['Thousands of Passengers','WMA6','WMA12']].head()
```

```python
airline[['Thousands of Passengers','WMA6','WMA12']].plot()
```

### Simple Exponential Smoothing, Holt's, Holt-Winter's

```python
from statsmodels.tsa.api import SimpleExpSmoothing
```

```python
train=airline[:'1959']
test=airline['1959':]
```

```python
train['Thousands of Passengers'].plot(figsize=(12,6))
test['Thousands of Passengers'].plot()
```

```python
ses_model=SimpleExpSmoothing(np.array(train['Thousands of Passengers']))
```

```python
ses_result=ses_model.fit()
```

```python
y_hat=test.copy()
```

```python
y_hat['SES']=ses_result.forecast(len(test))
```

```python
plt.figure(figsize=(15,8))
plt.plot(train['Thousands of Passengers'], label='Train')
plt.plot(test['Thousands of Passengers'], label='Test')
plt.plot(y_hat['SES'], label='Simple Exp Smoothing')
plt.legend()
```

```python
rmse=np.sqrt(mean_squared_error(test['Thousands of Passengers'],y_hat['SES']))
```

```python
rmse
```

### Holt's

```python
from statsmodels.tsa.api import Holt
```

```python
holt_model=Holt(np.asarray(train['Thousands of Passengers']))
```

```python
holt_result=holt_model.fit()
```

```python
y_hat['HOLT']=holt_result.forecast(len(test))
```

```python
plt.figure(figsize=(15,8))
plt.plot(train['Thousands of Passengers'], label='Train')
plt.plot(test['Thousands of Passengers'], label='Test')
plt.plot(y_hat['HOLT'], label='Holts Exp Smoothing')
plt.legend()
```

```python
rmse=np.sqrt(mean_squared_error(test['Thousands of Passengers'],y_hat['HOLT']))
```

```python
rmse
```

### Holt-Winters

```python
from statsmodels.tsa.api import ExponentialSmoothing
```

```python
winter_model=ExponentialSmoothing(np.array(train['Thousands of Passengers']), seasonal_periods=12, trend='add', seasonal='add')
```

```python
winter_result=winter_model.fit()
```

```python
y_hat['WINTER']=winter_result.forecast(len(test))
```

```python
plt.figure(figsize=(15,8))
plt.plot(train['Thousands of Passengers'], label='Train')
plt.plot(test['Thousands of Passengers'], label='Test')
plt.plot(y_hat['WINTER'], label='Holt-Winters Exp Smoothing')
plt.legend()
```

```python
rmse=np.sqrt(mean_squared_error(test['Thousands of Passengers'], y_hat['WINTER']))
```

```python
rmse
```

### ARIMA

```python
import statsmodels.api as sm
```

ARIMA(p,d,q)(P,D,Q)m

```python
arima=sm.tsa.statespace.SARIMAX(train['Thousands of Passengers'],
                                order=(2,1,1),
                                seasonal_order=(0,1,0,12),
                               enforce_stationarity=False,
                               enforce_invertibility=False)
```

```python
arima_result=arima.fit()
```

```python
y_hat['ARIMA']=\
arima_result.predict(start='1960-01-01',end='1960-12-01', dynamic=True)
```

```python
plt.figure(figsize=(15,8))
plt.plot(train['Thousands of Passengers'], label='Train')
plt.plot(test['Thousands of Passengers'], label='Test')
plt.plot(y_hat['ARIMA'], label='Seasonal ARIMA')
plt.legend()
```

```python
rmse=np.sqrt(mean_squared_error(test['Thousands of Passengers']['1960'], y_hat['ARIMA']['1960']))
```

```python
rmse
```
