## Rolling and Expanding

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df=pd.read_csv('./data/apple_stock.csv', index_col='Date', parse_dates=True)
```

```python
df.index
```

```python
def dateparser(str_dt):
    return pd.datetime.strptime(str_dt, '%Y-%m-%d')
```

```python
df=pd.read_csv('./data/apple_stock.csv', index_col='Date', parse_dates=['Date'], date_parser=dateparser)
```

```python
df.rolling(7).mean().head(10)
```

```python
df['2018':]['Close'].plot()
df['2018':].rolling(window=30).mean()['Close'].plot()
```

```python
df['Close: 30 Day Mean']=df['Close'].rolling(30).mean()
```

```python
df[['Close','Close: 30 Day Mean']].plot(figsize=(15,7))
```

```python
list(range(10))
```

```python
df['Close'].expanding(min_periods=1).mean().plot()
```

```python
df['Close : 20Day Mean']=df['Close'].rolling(20).mean()
```

```python
df['Upper']=df['Close : 20Day Mean']+2*df['Close'].rolling(20).std()
```

```python
df['Lower']=df['Close : 20Day Mean']-2*df['Close'].rolling(20).std()
```

```python
df['2018':][['Close','Close : 20Day Mean','Upper','Lower']].plot(figsize=(15,7))
```
