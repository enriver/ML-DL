## Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df=pd.read_csv('./data/apple_stock.csv', index_col='Date', parse_dates=True)
```

```python
df.head()
```

```python
df[['Volume', 'Adj Close']].plot(secondary_y='Volume')
```

```python
df['Adj Close'].plot()
```

```python
df['Adj Close'].plot(xlim=['2015-01-01','2018-01-01'], ylim=[70,180])
```

```python
import matplotlib.dates as dates
```

```python
index=df['2015-01-01':'2018-01-01'].index
stock=df['2015-01-01':'2018-01-01']['Adj Close']

fig,ax=plt.subplots(figsize=(15,3))
ax.plot_date(index,stock,'-')
ax.xaxis.set_major_locator(dates.MonthLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('%b\n%Y'))
fig.autofmt_xdate()
plt.tight_layout()
```
