# 다항 회귀 (Polynomial Regression)

```python
from sklearn.preprocessing import PolynomialFeatures
import numpy as np
```

```python
X=np.arange(4).reshape(2,2)

print('일차 단항식 계수 피처 : \n',X)
```

```python
poly=PolynomialFeatures(degree=2)
poly.fit(X)
X_poly=poly.transform(X)

print('변환된 2차 다항식 계수 피처 : \n',X_poly)
```

$y=x^3+x^2+2x+1$

```python
def polynomial_func(x):
    y=x**3+x**2+2*x+1
    return y
```

```python
y=polynomial_func(X)
print('삼차 다항식 결정값 : \n', y)
```

일차 단항식 계수를 삼차 다항식 계수로 변환하고, 이를 이용해 선형 회귀에 적용하여 다항 회귀 구현

```python
X_degree_3=PolynomialFeatures(degree=3).fit_transform(X)
print('3차 다항식 계수 feature : \n', X_degree_3)
```

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()
model.fit(X_degree_3, y)
```

```python
print('Polynomial 회귀 계수 \n', np.round(model.coef_ , 2))
print('Polynomial 회귀 Shape \n', model.coef_.shape)
```

### 과대/과소 적합

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
%matplotlib inline
```

```python
def cos_fun(x):
    return np.cos(1.5 * np.pi * x)
```

```python
np.random.seed(0)

n_samples=30
X=np.sort(np.random.rand(n_samples))

y=cos_fun(X)+np.random.randn(n_samples)*0.1
```
