#### 이상치 제거

```python
def find_outliers(x):
    Q1=np.percentile(x,25)
    Q3=np.percentile(x,75)
    
    IQR=Q3-Q1
    lower=Q1-1.5*IQR
    upper=Q3+1.5*IQR
    
    return np.where((x>upper)|(x<lower))
```

```python
outlier_index=set()

# distplot 으로 보았을때, 이상치의 영향이 컸던 변수들의 이상치 제거

for i in ['Y_Maximum','Y_Minimum','Outside_X_Index']:
    outlier_index.update(find_outliers(data[i])[0][:])
outlier_index=list(outlier_index)
```

```python
new_data=data.drop(outlier_index).reset_index()
```

```python
new_data.shape
```

이상치를 제거한 결과

```python
figure,axes=plt.subplots(ncols=3, figsize=(12,5))

sns.distplot(new_data['Y_Maximum'],ax=axes[0])
sns.distplot(new_data['Y_Minimum'],ax=axes[1])
sns.distplot(new_data['Outside_X_Index'],ax=axes[2])

plt.show()
```
