#### 이상치 제거

```python
def find_outliers(x):
    Q1=np.percentile(x,25)
    Q3=np.percentile(x,75)
    
    IQR=Q3-Q1
    lower=Q1-1.5*IQR
    upper=Q3+1.5*IQR
    
    return np.where((x>upper)|(x<lower))
```

```python
# boxplot 상 IQR 외부에 있던 변수들의 이상치 제거

outlier_index=set()

for i in numeric:
    outlier_index.update(find_outliers(df[i])[0][:])
    
outlier_index=list(outlier_index)
```

```python
new_df=df.drop(outlier_index).reset_index()
```

#### ① 첫번째 제공 파일의 총사용량 컬럼을 용도별로 분류하고, 연월과 사용 목적별로 전력의 하루 평균 사용량을 구하여 도표를 도출하시오.

|YYYYMM|A|B|C|D|E|
|-|-|-|-|-|-|
|202001|__|__|__|__|__|
|202002|__|__|__|__|__|
|202003|__|__|__|__|__|


##### CREATE DIFF FROM CUMSUM

```python
history['diff_a']=history['A'].diff().fillna(0)
history['diff_b']=history['B'].diff().fillna(0)
history['diff_c']=history['C'].diff().fillna(0)
history['diff_d']=history['D'].diff().fillna(0)
history['diff_e']=history['E'].diff().fillna(0)

history.head(10)
```

##### CREATE JOIN KEY

```python
usage['timestamp']=pd.date_range('09/01/2017', periods=6720, freq='900s')
usage['timestamp']=usage['timestamp'].apply(lambda x:x.date())

usage.head()
```

```python
t_usage=usage.groupby('timestamp')['amount'].sum()

total_usage=pd.DataFrame()
total_usage['timestamp']=t_usage.index
total_usage['total_amount']=t_usage.values

total_usage.head()
```

```python
history['time']=pd.date_range('09/01/2017', periods=201600, freq='30s')
history['time']=history['time'].apply(lambda x:x.date())

history.head()
```

```python
diff=history.groupby('time')[['diff_a','diff_b','diff_c','diff_d','diff_e']].sum()

diff['time']=diff.index
diff=diff.reset_index(drop=True)

diff.columns=['A','B','C','D','E','time']

diff.head()
```

```python
weather.dt=weather.dt.apply(lambda x: pd.to_datetime(x, format='%Y-%m-%d').date())
weather.head()
```

##### JOIN DATASETS

```python
electric=total_usage.merge(diff, left_on='timestamp', right_on='time')
```

```python
electric=electric.merge(weather, left_on='timestamp', right_on='dt')

electric.head()
```

```python
electric=electric.drop(['dt','time'], axis=1)
electric.head()
```

```python
electric['year']=electric['timestamp'].apply(lambda x : x.year)
electric['month']=electric['timestamp'].apply(lambda x : '%02d' %x.month)
electric['day']=electric['timestamp'].apply(lambda x: x.day)
electric['day_name']=electric['timestamp'].apply(lambda x: x.weekday())

electric.head()
```

```python
electric['YYYYMM']=electric['year'].astype(str)+electric['month'].astype(str)

electric.head()
```

```python
electric.groupby('YYYYMM')[['A','B','C','D','E']].mean()
```

#### ② 요일별 평균 전력사용량을 도출하시오. 또한 가로축을 요일, 세로축을 평균사용량으로 하여 요일별 평균 사용량을 시각화하여 제출하시오.

```python
electric['day_name']=electric['day_name'].replace({0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'})
electric['day_name'].unique()
```

```python
day_amount=electric.groupby('day_name')['total_amount'].mean()
day_amount
```

#### ③ 요일별 총 전력 사용량의 평균값의 차이를 분석하여, 가장 큰 차이를 보이는 요일은 어떤 요일인지 제시하시오.

```python
max_day=day_amount.values.max()
min_day=day_amount.values.min()

print('제일 많이 사용하는 요일 : ',day_amount.index[np.argmax(day_amount.values)])
print('제일 적게 사용하는 요일 : ',day_amount.index[np.argmin(day_amount.values)])
```
